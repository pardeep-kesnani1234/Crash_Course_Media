```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Solution for Assignment #2

```python
img_detection = cv2.imread('john1.jpg')

# Initilizing the haar classifier with the face detector model
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')  

# perform the detection , here we are using 1.3 scale factor and 5 min neighbours
faces = face_cascade.detectMultiScale(img_detection, 1.3, 5)

#initilize the eye Cascade:
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

# loop through each face and draw it
for (x,y,w,h) in faces:
    # cv2.rectangle(img_detection,(x,y),(x+w,y+h),(0,255,255),4)

    # Extract roi of the face
    roi = img_detection[y:y+h, x:x+w]
    
    #perform eye detection with the default parameters i.e. 1.1 , 3
    eyes = eye_cascade.detectMultiScale(roi)  
    
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

plt.figure(figsize=[13,13])
plt.imshow(img_detection[:,:,::-1]);plt.axis("off");
```

```python
# Save the Output
cv2.imwrite( 'outputs/eyedetected.jpg', img_detection)
```

## Solution For Assignment #1

```python
# Read Image
img = cv2.imread('naruto.JPG')

# Reat the cat image
cat_face = cv2.imread('catface.JPG')

# Resize the cat
cat_face_resized = cv2.resize(cat_face, (150, 170))

# Grab the Face ROI(Region of Interest)
img[100:270,300:450]  =  cat_face_resized 

# Show the image
plt.figure(figsize=[13,13])
plt.imshow(img[:,:,::-1]);plt.axis("off");
```

```python
# Save the Output
cv2.imwrite( 'outputs/narutocat.jpg', img)
```

## Solution for Assignment 3

```python
img = cv2.imread('object.jpg')

image_copy   = img.copy()

# Convert the color image to grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 

# Apply the threshold
_ , thresholded_image = cv2.threshold(gray_image, 220, 255, cv2.THRESH_BINARY_INV)

kernel = np.ones((7,7),np.uint8)

thresholded_image  = cv2.dilate(thresholded_image, kernel,iterations = 2)

# Detect Contours
contours, _ = cv2.findContours(thresholded_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

print("Length of Contours is: {}".format(len(contours)))
# Draw the detected contours, -1 means draw all detected contours
cv2.drawContours(image_copy , contours, -1 , (0,255,0), 3)

# Show the image
plt.figure(figsize=[13,13])
plt.imshow(image_copy[:,:,::-1]);plt.axis("off");
```

```python
# Save the Output
cv2.imwrite( 'outputs/contours.jpg', image_copy)
```
